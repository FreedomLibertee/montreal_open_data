# Geolocation

##### Imports

```python
import pandas as pd
import numpy as np
import time

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
```

##### Paths

```python
filepath = 'Z:/Open_data_sets/Montreal_open_data/evaluation_fonciere/uniteevaluationfonciere.csv'
```

Note: Download the data file from the Montreal Open Data Portal here:
        http://donnees.ville.montreal.qc.ca/dataset/unites-evaluation-fonciere/resource/2b9dfc3d-91d3-48de-b32c-a2a6d9417079

```python
# Load as dataframe
df = pd.read_csv(filepath,  sep = ',') #+|\t+|\s+\t+|\t+\s+')        
df.sort_values(by='ANNEE_CONSTRUCTION', inplace=True)
df.drop_duplicates(subset=['CIVIQUE_DEBUT', 'NOM_RUE'], inplace=True)
display(df)
```

##### Define a subset (Rosemont) - skip this to process entire dataframe

```python
df = df.loc[df['NO_ARROND_ILE_CUM'] == "REM24"]
print(len(df))
display(df)
```

##### Run the geolocator on the list of addresses

```python
geolocator = Nominatim()
geolocator = Nominatim(format_string="%s, Montreal QC")
array = np.empty([len(df),4])
name_list = []
j = 0;
for i in range (0,len(df)):
    loc_str = str(df.iloc[i]['CIVIQUE_DEBUT']) + ' ' + str(df.iloc[i]['NOM_RUE'])
    loc_str = loc_str[:loc_str.find('(')]
    
#     # A few cases of exceptional spelling that need to be addressed
#     if 'croissant' in loc_str:
#         loc_str = loc_str.replace('croissant', 'cres')
#     if " l' " in loc_str:
#         loc_str = loc_str.replace("l' ", "l'")
    
    try:
#         print(loc_str)
        time.sleep(1)
        location = geolocator.geocode(loc_str, timeout=10)
    except GeocoderTimedOut as e:
        print("Error: geocode failed on input %s "%(loc_str))
    if location is not None:
        array[j,0] = location.longitude
        array[j,1] = location.latitude
        array[j,2] = df.iloc[i]['ANNEE_CONSTRUCTION']
        array[j,3] = df.iloc[i]['SUPERFICIE_BATIMENT']
        name_list.append(loc_str)
        print(str(i),str(j),name_list[j], array[j,:])
#     else: 
#         print("Skipping: i= " + str(i) + ", address: " + loc_str)
#         j = j - 1 # Stop us from advancing j if an address is skipped.
#     j = j + 1
```
